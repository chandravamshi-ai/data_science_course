# Simple Linear Regression

## y = ax +b
 
 * The above equation represents simple linear regression formula a.k.a equation of straight line
 * Where 
  * y is dependent variable
  * x is independent variable
  * b is intercept also called constant. b value is easily determiened i.e.., point where slope crosses y axis is b value
  * a is slope or coefficient can be calculated by using formulae (unit change in y / unit change in x)
  * correlation is used to find whether slope is + or - or neutral
 

### Example
simple linear regression with example

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Salary_Data.csv")
data.head()
```

```python
len(data)
```

#### The Goal here is to predict the salary based on the experience
* Our data consist of two columns and 30 rows YearsExperienece which is experience. It can be considered as x varaible i.e.., independent
* Salary is dependednt varialbe y
* Our linear equation is Salary (y) = a * YearsExperience(x) + b

```python
X = data.iloc[:,0:1].values
# x contains only YearsExperience data (independent variable) also called features
y = data.iloc[:,-1].values
# y contains only Salary data (dependent varialbe)
```

```python
x.shape,y.shape
# one thing here to note is 
# x is martrix of features (independent varialble).
# y is vector of dependent variable
```

##### Divide the data into train set  and test set
* The data is divided into two parts training data and testing data.
* A good split is dividing 70% of the data into training. Because the more diversified data you train the more model will learn.
* 30% Testing data
* I will use sklearn to split the data

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

##### Fitting Simple Linear Regression to the Training set

```python
from sklearn.linear_model import LinearRegression
#from linear_model library importing LinerarRegression class and creating object of LinearRegression 
# use fit method to fit model LinearRegression on the training data
regressor = LinearRegression()
# regressor is object of class LinearRegression
regressor.fit(X_train, y_train)
```

##### Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
# we use predict function to predict on the new values or to make predictions on untrained data
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,7),dpi=150)

ax[0].scatter(X_test,y_test, color='g',label='Actual values')
ax[0].plot(X_train, regressor.predict(X_train), color = 'r',label='Predicted  values')
ax[0].set_title('Actual vs Prected values on test set')
ax[0].set_xlabel('Experience')
ax[0].set_ylabel('Salary')
ax[0].legend(loc='best')


ax[1].scatter(X_train,y_train, color='green',label='Actual values')
ax[1].plot(X_train, regressor.predict(X_train), color = 'r',label='Predicted  values')
ax[1].set_title('Actual vs Prected values on train set')
ax[1].set_xlabel('Experience')
ax[1].set_ylabel('Salary')
ax[1].legend(loc='best')

plt.show()
```
